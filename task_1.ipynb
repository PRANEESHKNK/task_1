```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\praneesh\Desktop\DATASCIENCE\cities_india.csv')
df.head()
df.tail()
df.shape
print(df.columns)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['population_male'], kde=True)
plt.xlabel('population_male')
plt.ylabel('Frequency')
plt.title('population_male')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['population_female'], kde=True)
plt.xlabel('population_female')
plt.ylabel('Frequency')
plt.title('population_female')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=df['state_name'], y=df['population_male'], color='red')
plt.xlabel('state_name')
plt.ylabel('population_male')
plt.xticks(rotation=45, ha='right')
plt.title('population of male in every state')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=df['population_male'], y=df['population_female'])
plt.xlabel('population_male')
plt.ylabel('population_female')
plt.title('population_female vs population_male')
plt.show()
```
